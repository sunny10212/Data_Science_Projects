```python
import sys
```

```python
def unit_translator(filename, D):
     
       
        #we have defined the function above. In this function, required file has to be mentioned along with type. If there is a notepad file named "conversion",mention it as "conversion.txt"."""
    # Mention the filename and define dictonary by mentioning the units to be shown as output."""
      
    
    L = {"mm": 10**6, "cm": 10**5, "m": 10**3, "km": 10**0};
    F = {"B": 1024**5, "KB": 1024**4, "MB": 1024**3, "GB": 1024**2, "TB": 1024**1, "PB": 1024**0};
    T = {"sec": 3600, "min": 60, "h":1 };
     #dictonaries have been created above.
    #file_name is a variable to open the file"""
    file_name = open('test.txt','r');

    #Here "content" is a variable used to read the body of the file."""
    content = file_name.read();

    #Here we are creating a word list just by using wor."""
    wor = content.split( );

    #this 'enumerate' unlike the other function is used to track the position of any item. Here we are using it for performing repetition.
    
    for p,q in enumerate(wor):

        dott = False;

    #Here the '.' is named as "dott". This condition checks and searches for the '.' is the content or any specific line.
    # If dott is True then element 'q' will be identify the unit just before the '.'"".
    #wor[p] = q means that whichever unit is there before the '.', that unit will be assisgned to element q for conversion
        if q[-1] == '.' :
            q = q[:-1];
                      
            wor[p] = q;
            dott = True;

# Now we are converting the units of length by identifying length,filesize and time in dictonary keys and element in corresponding keys.
    
        if 'L' in D.keys() and q in L.keys():
            
# Once the unit is assigned to element q, then the conversions take place by storing the given numeric value and unit in a temprory variable. And conversion is done by using isnumeric().
        
            temprory = wor[p-1];
            temprory = temprory.replace('.','').replace(',','').replace('!','').replace(' .','').replace(' ,','').replace('. ',' ').replace(', ',' ').replace('! ',' ');
            if not temprory.isnumeric():
                continue;
            
            if wor[p-1].isnumeric():
                
                x = int(wor[p-1]);
            else:
                x = float(wor[p-1]);
        
        
            x = x * (L[D["L"]] / L[q])
            
            
            wor[p] = D["L"];
            wor[p-1] = float(x);

        if 'F' in D.keys() and q in F.keys():

            temprory = wor[p-1];
            temprory = temprory.replace('.','').replace(',','').replace('!','').replace(' .','').replace(' ,','').replace('. ',' ').replace(', ',' ').replace('! ',' ');
           
            if not temprory.isnumeric():
            	continue;
            	
            if wor[p-1].isnumeric():
                x = int(wor[p-1]);
            else:
                x = float(wor[p-1]);

            
            x = x * (F[D["F"]] / F[q]);

            
            wor[p] = D["F"];
            wor[p-1] = float(x);

        if 'T' in D.keys() and q in T.keys():

            temprory = wor[p-1];
            temprory = temprory.replace('.','').replace(',','').replace('!','').replace(' .','').replace(' ,','').replace('. ',' ').replace(', ',' ').replace('! ',' ');
            if not temprory.isnumeric():
            	continue;
            if wor[p-1].isnumeric():
    
                x = int(wor[p-1]);
            else:
                x = float(wor[p-1]);

            
            x = x * (T[D["T"]] / T[q]);

           
            wor[p] = D["T"];
            wor[p-1] = float(x);

        if dott == '.' :
            wor[p] = wor[p] + '.';
        if dott == ',' :
            wor[p] = wor[p] + ',';
        if dott == '!' :
            wor[p] = wor[p] + '!';
 
            

# Now the file is to be closed using file_name.close()    
    file_name.close();

#As the output will be shown in new file where last four letters of the filename gets replaced by "_translated.txt"    
    newfile_name =filename[:-4] + "_translated.txt";

#we will now open the new file using the same variable file_name. This time, we are going to write the file so that conversions can be shown in the newfile.    
    file_name = open(newfile_name,"w");

    
    for w in wor:
        file_name.write(str(w));
        file_name.write(" ");

    
    file_name.close();
    


```

```python
# That's how the dictionary needs to be defined
D = {'L': 'cm', 'F': 'GB'}

# ---- DO NOT CHANGE THE CODE BELOW ----
if __name__ == "__main__":
    if len(sys.argv)<2: raise ValueError('Provide file_namename as input argument')
    file_namename = sys.argv[1]
    print('file_namename is "{}"'.format(file_namename))
    unit_translator(file_namename, D)
```
